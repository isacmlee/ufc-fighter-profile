```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
fighter_profile_url = "https://www.ufc.com/athlete/ilia-topuria"
```

```python
# Send a GET request to the page
headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(fighter_profile_url, headers=headers)

# Parse the HTML
soup = BeautifulSoup(response.text, "html.parser")
```

```python
fighter_name = soup.find("h1", class_="hero-profile__name").text
fighter_nickname = soup.find("p", class_="hero-profile__nickname").text
weight_division = soup.find("p", class_="hero-profile__division-title").text
record = soup.find("p", class_="hero-profile__division-body").text
knockout_wins = soup.find_all("p", class_="hero-profile__stat-numb")[0].text
submission_wins = soup.find_all("p", class_="hero-profile__stat-numb")[1].text
first_round_finishes = soup.find_all("p", class_="hero-profile__stat-numb")[2].text
striking_accuracy_percentage = soup.find_all("text", class_="e-chart-circle__percent")[0].text
sig_strikes_landed = soup.find_all("dd", class_="c-overlap__stats-value")[0].text
sig_strikes_attempted = soup.find_all("dd", class_="c-overlap__stats-value")[1].text
takedown_accuracy_percentage = soup.find_all("text", class_="e-chart-circle__percent")[1].text
takedowns_landed = soup.find_all("dd", class_="c-overlap__stats-value")[2].text
takedowns_attempted = soup.find_all("dd", class_="c-overlap__stats-value")[3].text
sig_strikes_landed_per_min = soup.find_all("div", class_="c-stat-compare__number")[0].text
sig_strikes_absorbed_per_min = soup.find_all("div", class_="c-stat-compare__number")[1].text
takedown_avg_per_15_min = soup.find_all("div", class_="c-stat-compare__number")[2].text
submission_avg_per_15_min = soup.find_all("div", class_="c-stat-compare__number")[3].text
sig_strike_defense = soup.find_all("div", class_="c-stat-compare__number")[4].text
takedown_defense = soup.find_all("div", class_="c-stat-compare__number")[5].text
knockdown_avg_per_15_min = soup.find_all("div", class_="c-stat-compare__number")[6].text
avg_fight_time = soup.find_all("div", class_="c-stat-compare__number")[7].text
sig_strike_by_standing = soup.find_all("div", class_="c-stat-3bar__value")[0].text
sig_strike_by_clinch = soup.find_all("div", class_="c-stat-3bar__value")[1].text
sig_strike_by_ground = soup.find_all("div", class_="c-stat-3bar__value")[2].text
win_by_ko_tko = soup.find_all("div", class_="c-stat-3bar__value")[3].text
win_by_decision = soup.find_all("div", class_="c-stat-3bar__value")[4].text
win_by_sub = soup.find_all("div", class_="c-stat-3bar__value")[5].text
sig_strike_by_head_percent = soup.find(id="e-stat-body_x5F__x5F_head_percent").text
sig_strike_by_body_percent = soup.find(id="e-stat-body_x5F__x5F_body_percent").text
sig_strike_by_leg_percent = soup.find(id="e-stat-body_x5F__x5F_leg_percent").text
```

```python
data = {
    "fighter_name": fighter_name,
    "figher_nickname": fighter_nickname,
    "weight_division": weight_division,
    "record": record,
    "knockout_wins": knockout_wins,
    "submission_wins": submission_wins,
    "first_round_finishes": first_round_finishes,
    "striking_accuracy_percentage": striking_accuracy_percentage,
    "sig_strikes_landed": sig_strikes_landed,
    "sig_strikes_attempted": sig_strikes_attempted,
    "takedown_accuracy_percentage": takedown_accuracy_percentage,
    "takedowns_landed": takedowns_landed,
    "takedowns_attempted": takedowns_attempted,
    "sig_strikes_landed_per_min": sig_strikes_landed_per_min,
    "sig_strikes_absorbed_per_min": sig_strikes_absorbed_per_min,
    "takedown_avg_per_15_min": takedown_avg_per_15_min,
    "submission_avg_per_15_min": submission_avg_per_15_min,
    "sig_strike_defense": sig_strike_defense,
    "takedown_defense": takedown_defense,
    "knockdown_avg_per_15_min": knockdown_avg_per_15_min,
    "avg_fight_time": avg_fight_time,
    "sig_strike_by_standing": sig_strike_by_standing,
    "sig_strike_by_clinch": sig_strike_by_clinch,
    "sig_strike_by_ground": sig_strike_by_ground,
    "win_by_ko_tko": win_by_ko_tko,
    "win_by_decision": win_by_decision,
    "win_by_sub": win_by_sub,
    "sig_strike_by_head_percent": sig_strike_by_head_percent,
    "sig_strike_by_body_percent": sig_strike_by_body_percent,
    "sig_strike_by_leg_percent": sig_strike_by_leg_percent
}
```

```python
df = pd.DataFrame([data])
```
