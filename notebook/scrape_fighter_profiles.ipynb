```python
# Load UFC fighters search page
base_url = "https://www.ufc.com/search?query=&tabOrder=.%2Findex.html%2Cathletes%2Chelp_articles%2Cvideos%2Cfaqs%2Clinks%2Cevents%2Cpromotions&facetFilters=%7B%22gender%22%3A%5B%7B%22gender%22%3A%7B%22%24eq%22%3A%22Male%22%7D%7D%5D%2C%22c_weightClass%22%3A%5B%5D%2C%22c_homeCountry%22%3A%5B%5D%2C%22c_accolades%22%3A%5B%5D%7D&filters=%7B%7D&search-offset={}&referrerPageUrl=&verticalUrl=Athletes.html"
offset = 0
step = 21  
fighter_profile_urls = []

while True: 
    print(offset)
    url = base_url.format(offset)
    temp_urls = extract_profile_urls(url)

    if temp_urls is None:
        break
    else:
        time.sleep(5)
        fighter_profile_urls.append(temp_urls)
        offset+=step
```
