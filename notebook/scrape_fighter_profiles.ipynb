```python
def find_element_text_or_none(soup, target_element, class_name): 
    element = soup.find(target_element, class_=class_name)
    return element.text if element else None

def find_all_elements_or_none(soup, target_element, class_name):
    list_of_elements = soup.find_all(target_element, class_=class_name)
    return list_of_elements

def find_element_text_or_none_id(soup, id):
    element = soup.find(id=id)
    return element.text if element else None

def scrape_fighter_stats(url):
    # Send a GET request to the page
    headers = {"User-Agent": "Mozilla/5.0"}
    response = requests.get(fighter_profile_url, headers=headers)
    
    # Parse the HTML
    soup = BeautifulSoup(response.text, "html.parser")

    fighter_name = find_element_text_or_none(soup,"h1", "hero-profile__name")
    fighter_nickname = find_element_text_or_none(soup, "p", "hero-profile__nickname")
    weight_division = find_element_text_or_none(soup, "p", "hero-profile__division-title")
    record = find_element_text_or_none(soup, "p", "hero-profile__division-body")
    knockout_wins = find_all_elements_or_none(soup, "p", "hero-profile__stat-numb")[0].text
    submission_wins = find_all_elements_or_none(soup, "p", "hero-profile__stat-numb")[1].text
    first_round_finishes = find_all_elements_or_none(soup, "p", "hero-profile__stat-numb")[2].text
    striking_accuracy_percentage = find_all_elements_or_none(soup, "text", "e-chart-circle__percent")[0].text
    sig_strikes_landed = find_all_elements_or_none(soup, "dd", "c-overlap__stats-value")[0].text
    sig_strikes_attempted = find_all_elements_or_none(soup, "dd", "c-overlap__stats-value")[1].text
    takedown_accuracy_percentage = find_all_elements_or_none(soup, "text", "e-chart-circle__percent")[1].text
    takedowns_landed = find_all_elements_or_none(soup, "dd", "c-overlap__stats-value")[2].text
    takedowns_attempted = find_all_elements_or_none(soup, "dd", "c-overlap__stats-value")[3].text
    sig_strikes_landed_per_min = find_all_elements_or_none(soup, "div", "c-stat-compare__number")[0].text
    sig_strikes_absorbed_per_min = find_all_elements_or_none(soup, "div", "c-stat-compare__number")[1].text
    takedown_avg_per_15_min = find_all_elements_or_none(soup, "div", "c-stat-compare__number")[2].text
    submission_avg_per_15_min = find_all_elements_or_none(soup, "div", "c-stat-compare__number")[3].text
    sig_strike_defense = find_all_elements_or_none(soup, "div", "c-stat-compare__number")[4].text
    takedown_defense = find_all_elements_or_none(soup, "div", "c-stat-compare__number")[5].text
    knockdown_avg_per_15_min = find_all_elements_or_none(soup, "div", "c-stat-compare__number")[6].text
    avg_fight_time = find_all_elements_or_none(soup, "div", "c-stat-compare__number")[7].text
    sig_strike_by_standing = find_all_elements_or_none(soup, "div", "c-stat-3bar__value")[0].text
    sig_strike_by_clinch = find_all_elements_or_none(soup, "div", "c-stat-3bar__value")[1].text
    sig_strike_by_ground = find_all_elements_or_none(soup, "div", "c-stat-3bar__value")[2].text
    win_by_ko_tko = find_all_elements_or_none(soup, "div", "c-stat-3bar__value")[3].text
    win_by_decision = find_all_elements_or_none(soup, "div", "c-stat-3bar__value")[4].text
    win_by_sub = find_all_elements_or_none(soup, "div", "c-stat-3bar__value")[5].text
    sig_strike_by_head_percent = find_element_text_or_none_id(soup, "e-stat-body_x5F__x5F_head_percent")
    sig_strike_by_body_percent = find_element_text_or_none_id(soup, "e-stat-body_x5F__x5F_body_percent")
    sig_strike_by_leg_percent = find_element_text_or_none_id(soup, "e-stat-body_x5F__x5F_leg_percent")
    status = find_all_elements_or_none(soup, "div", "c-bio__text")[0].text
    place_of_birth = find_all_elements_or_none(soup, "div", "c-bio__text")[1].text
    fighting_style = find_all_elements_or_none(soup, "div", "c-bio__text")[2].text
    age = find_all_elements_or_none(soup, "div", "c-bio__text")[3].text
    height = find_all_elements_or_none(soup, "div", "c-bio__text")[4].text
    weight = find_all_elements_or_none(soup, "div", "c-bio__text")[5].text
    octagon_debut = find_all_elements_or_none(soup, "div", "c-bio__text")[6].text
    reach = find_all_elements_or_none(soup, "div", "c-bio__text")[7].text
    leg_reach = find_all_elements_or_none(soup, "div", "c-bio__text")[8].text

    data = {
        "fighter_name": fighter_name,
        "figher_nickname": fighter_nickname,
        "weight_division": weight_division,
        "record": record,
        "knockout_wins": knockout_wins,
        "submission_wins": submission_wins,
        "first_round_finishes": first_round_finishes,
        "striking_accuracy_percentage": striking_accuracy_percentage,
        "sig_strikes_landed": sig_strikes_landed,
        "sig_strikes_attempted": sig_strikes_attempted,
        "takedown_accuracy_percentage": takedown_accuracy_percentage,
        "takedowns_landed": takedowns_landed,
        "takedowns_attempted": takedowns_attempted,
        "sig_strikes_landed_per_min": sig_strikes_landed_per_min,
        "sig_strikes_absorbed_per_min": sig_strikes_absorbed_per_min,
        "takedown_avg_per_15_min": takedown_avg_per_15_min,
        "submission_avg_per_15_min": submission_avg_per_15_min,
        "sig_strike_defense": sig_strike_defense,
        "takedown_defense": takedown_defense,
        "knockdown_avg_per_15_min": knockdown_avg_per_15_min,
        "avg_fight_time": avg_fight_time,
        "sig_strike_by_standing": sig_strike_by_standing,
        "sig_strike_by_clinch": sig_strike_by_clinch,
        "sig_strike_by_ground": sig_strike_by_ground,
        "win_by_ko_tko": win_by_ko_tko,
        "win_by_decision": win_by_decision,
        "win_by_sub": win_by_sub,
        "sig_strike_by_head_percent": sig_strike_by_head_percent,
        "sig_strike_by_body_percent": sig_strike_by_body_percent,
        "sig_strike_by_leg_percent": sig_strike_by_leg_percent,
        "status": status,
        "place_of_birth": place_of_birth,
        "fighting_style": fighting_style,
        "age": age,
        "height": height,
        "weight": weight,
        "octagon_debut": octagon_debut,
        "reach": reach,
        "leg_reach": leg_reach
    }

    df = pd.DataFrame([data])

    return df 
```
